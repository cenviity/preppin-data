```python
df = pd.read_csv("input/2024_01.csv")
df.head()
```

```python
# Split the `Flight Details` field to form new fields
split_data = df["Flight Details"].str.split("//", expand=True)
split_data.columns = ["Date", "Flight Number", "From-To", "Class", "Price"]

df = df.assign(**split_data)
df
```

```python
# Split the `From-To` field as well
split_data = df["From-To"].str.split("-", expand=True)
split_data.columns = ["From", "To"]

df = df.assign(**split_data)
df
```

```python
df.dtypes
```

```python
# Convert data types
df = df.astype(
    {
        "Date": "datetime64[ns]",
        "Price": float,
    }
)
df.dtypes
```

```python
df
```

```python
df["Flow Card?"] = df["Flow Card?"].replace({0: "No", 1: "Yes"})
df
```

```python
columns = [
    "Date",
    "Flight Number",
    "From",
    "To",
    "Class",
    "Price",
    "Flow Card?",
    "Bags Checked",
    "Meal Type",
]
df = df[columns]
df
```

```python
flow_card_groups = df.groupby(by="Flow Card?")

flow_card = flow_card_groups.get_group("Yes").reset_index(drop=True)
flow_card
```

```python
non_flow_card = flow_card_groups.get_group("No").reset_index(drop=True)
non_flow_card
```
